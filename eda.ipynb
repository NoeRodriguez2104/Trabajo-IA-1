```python
import pandas as pd
from plotnine import element_line, geom_smooth, ggplot, aes, geom_point, labs, theme

df = pd.read_csv('dataset\music_genre.csv')

plot = (
    ggplot(df, aes(x='energy', y='popularity')) +
    geom_point(color='blue', alpha=0.5) +
    geom_smooth(method='lm', color='red', se=False) +
    labs(title='Regresión lineal entre Energía y Popularidad',
        x='Energía', y='Popularidad') +
    theme(
        axis_line_x=element_line(color='black', size=0.5),
        axis_line_y=element_line(color='black', size=0.5)
    )
)
plot
```
